```python
import numpy
import os
```

```python
help(numpy.genfromtxt)
```

```python
distance_file = os.path.join('data', 'distance_data_headers.csv')

distances = numpy.genfromtxt(fname=distance_file, delimiter=',' , dtype='unicode')
print(distances)
```

```python
headers = distances[0]
print(headers)
```

```python
# Slice distances to get the numerical values (everything but not the header)
data = distances[1:]
print(data)
```

```python
data = data.astype(numpy.float)
print(data)
```

```python
# To slice a 2D
# array_name[row_index, column_index]

print(data[0,1])
print(data[1,0])
```

```python
# Ten rows three column, 0 could be removed as below too
small_data = data[0:10, 0:3]
print(small_data)
```

```python
# column without starting index
print(small_data[:, 1:])
```

```python
# data_average = numpy.mean(data_set)
```

```python
print(headers)
```

```python
thr4_atp = data[:,1]  # Every row, just the THR4_ATP column
avg_thr4_atp = numpy.mean(thr4_atp)
print(avg_thr4_atp)
```

```python
# Create a variable called num_columns which is the number of column in our data
num_columns = len(data[0,:]) # she used len(headers)
print(num_columns)
```

```python
for i in range(1, num_columns):
    #print(i)
    column = data[:, i]
    avg_col = numpy.mean(column)
    data_label = headers[i]
    print(F'{data_label} : {avg_col:.3f}')
```

## Plotting
Plotting data using matplotlib

```python
import matplotlib.pyplot as plt

## %matplotlib inline (if the above did not work)
```

```python
plt.figure()
plt.plot(data[:,1])
```

```python
plt.figure()

sample_name = headers[1]

plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:, 1], label=sample_name)
plt.legend()

plt.savefig(F'{sample_name}.png', dpi=300)
```

```python
plt.figure()


plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.plot(data[:, 1], label=headers[1], color='green')
plt.plot(data[:, 2], label=headers[2], color='black')
plt.legend()

plt.savefig('two_samples.png', dpi=300)
```

```python
plt.figure()

for col in range(1, num_columns):
    plt.plot(data[:,col], label=headers[col])
    
plt.xlabel('Simulation Frame')
plt.ylabel('Distance (angstrom)')
plt.legend()

plt.savefig('all_samples.png')
```

```python
# Make a plot for each sample
# Try adding savefig with file name equal to sample name

data_min = numpy.min(data[:,1:])
data_max = numpy.max(data[:,1:])

for col in range(1, num_columns):
    sample_name = headers[col]
    plt.figure()
    plt.plot(data[:, col], label=sample_name)
    plt.ylim()
    plt.xlabel('Simulation Frame')
    plt.ylabel('Distance (angstrom)')
    plt.legend()
    plt.savefig(F'{sample_name}.png')
```
